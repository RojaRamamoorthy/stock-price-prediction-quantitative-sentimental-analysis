```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

# Extract the feature data
X = df[feature_columns].values
# Extract the target variable
y = df['Close'].values
# Normalize the feature data using Min-Max scaling
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
# Reshape the input data for LSTM (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])
# Define the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(50, input_shape=(X_train.shape[1], X_train.shape[2])))
model_lstm.add(Dense(1))
model_lstm.compile(loss='mean_squared_error', optimizer='adam')
# Train the LSTM model
model_lstm.fit(X_train, y_train, epochs=50, batch_size=64, verbose=2)
# Make predictions on the test data
y_pred_lstm = model_lstm.predict(X_test)
# Calculate Mean Squared Error
mse_lstm = mean_squared_error(y_test, y_pred_lstm)
rmse_lstm = np.sqrt(mse_lstm)

print(f"LSTM Root Mean Squared Error (RMSE): {rmse_lstm}")
```

```python
import numpy as np
import pandas as pd
from keras.models import Sequential, Model
from keras.layers import LSTM, GRU, Dense, Input, concatenate
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
# Define your feature columns
feature_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Sentiment']

# Extract the feature data
X = df[feature_columns].values

# Extract the target variable
y = df['Close'].values

# Normalize the feature data using Min-Max scaling
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape the input data for LSTM and GRU (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])

# Define the input layer for both LSTM and GRU
input_layer = Input(shape=(X_train.shape[1], X_train.shape[2]))

# Define the LSTM model
lstm_layer = LSTM(50)(input_layer)
lstm_output = Dense(1)(lstm_layer)

# Define the GRU model
gru_layer = GRU(50)(input_layer)
gru_output = Dense(1)(gru_layer)

# Concatenate the outputs of LSTM and GRU
merged = concatenate([lstm_output, gru_output])

# Define the output layer for the combined model
output = Dense(1)(merged)

# Create the composite model
composite_model = Model(inputs=input_layer, outputs=output)

# Compile the composite model
composite_model.compile(loss='mean_squared_error', optimizer='adam')

# Train the composite model
composite_model.fit(X_train, y_train, epochs=50, batch_size=64, verbose=2)

# Make predictions on the test data
y_pred_combined = composite_model.predict(X_test)

# Calculate Mean Squared Error
mse_combined = mean_squared_error(y_test, y_pred_combined)


# Calculate RMSE
rmse_combined = np.sqrt(mse_combined)

print(f"Combined Model Root Mean Squared Error (RMSE): {rmse_combined}")
```

**MS-SSA-LSTM**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from scipy.linalg import hankel
from keras.models import Sequential
from keras.layers import LSTM, Dense
# Define the number of singular components to use
n_components = 5  # You can adjust this as needed
# Function to perform Multi-scale Singular Spectrum Analysis (MS-SSA)
def ms_ssa(series, n_components):
    L = len(series)
    hankel_matrix = hankel(series)
    U, s, Vt = np.linalg.svd(hankel_matrix)

    singular_vectors = []
    for i in range(n_components):
        singular_vectors.append(U[:, i].reshape(-1, 1))

    return singular_vectors
# Perform MS-SSA on the 'Close' column of your dataset
close_series = df['Close'].values
singular_vectors = ms_ssa(close_series, n_components)

# Combine the singular vectors into a feature matrix
feature_matrix = np.hstack(singular_vectors)

# Normalize the feature data using Min-Max scaling
scaler = MinMaxScaler()
feature_matrix = scaler.fit_transform(feature_matrix)

# Split the data into training and testing sets
train_size = int(len(feature_matrix) * 0.8)
X_train, X_test = feature_matrix[:train_size], feature_matrix[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape the input data for LSTM (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])

# Define the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(50, input_shape=(X_train.shape[1], X_train.shape[2])))
model_lstm.add(Dense(1))
model_lstm.compile(loss='mean_squared_error', optimizer='adam')

# Train the LSTM model
model_lstm.fit(X_train, y_train, epochs=50, batch_size=64, verbose=2)

# Make predictions on the test data
y_pred_lstm = model_lstm.predict(X_test)

# Calculate Mean Squared Error for the LSTM model
mse_lstm1 = mean_squared_error(y_test, y_pred_lstm)
rmse_lstm1 = np.sqrt(mse_lstm1)
print(f"MS-SSA-LSTM Root Mean Squared Error (RMSE): {rmse_lstm1}")
```
